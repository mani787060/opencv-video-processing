```python
import cv2
```

```python
# for capturing video :-

cap = cv2.VideoCapture(0)     # (source=0) -> shows that we are using webcam of our computer/laptop.

while True:
    ret, frame = cap.read()      # we have to read the frame from our webcam
    # ret(return)=true/false  and frame = actual image that captures at a particular moment

    if not ret:
        print("could not load image")
        break

    # to display the image which is read
    cv2.imshow("webcam feed", frame)

    # here waitKey(1) shows that we have to wait here for 1msec, to check if user presses any key from keyboard or not 
    # and if it presses key='q' then you have to exit from the loop and close the program
    if cv2.waitKey(1) & 0xFF==ord('q'):
        print("Quiting....")
        break

cap.release()                          # it closes our webcam
cv2.destroyAllWindows()                # it closes all windows of openCV (that is running)
```

```python
# saving webam video :-

camera = cv2.VideoCapture(0)    

# it captures the frames width and height of our webcam 
frame_width = int(camera.get(cv2.CAP_PROP_FRAME_WIDTH))        
frame_height = int(camera.get(cv2.CAP_PROP_FRAME_HEIGHT))   


codec = cv2.VideoWriter_fourcc(*'XVID')
recorder = cv2.VideoWriter("my_video.mp4", codec, 20, (frame_width, frame_height))


while True:
    success, image = camera.read()

    if not success:
        break

    recorder.write(image)
    cv2.imshow("Recording live", image)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

camera.release()
recorder.release()
cv2.destroyAllWindows()
```
